```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
df.describe()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
#selected_features = df[['names', 'of', 'selected', 'features', 'here']]
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y=df["koi_disposition"]
X=df.drop(columns=["koi_disposition"])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_scaler=MinMaxScaler().fit(X_train)
X_trains=X_scaler.transform(X_train)
X_tests=X_scaler.transform(X_test)
```

# Train the Model



```python
from sklearn.svm import SVC
model_svc=SVC(kernel="linear")
model_svc.fit(X_trains, y_train)
```

```python
print(f"Training Data Score: {model_svc.score(X_trains, y_train)}")
print(f"Testing Data Score: {model_svc.score(X_tests, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
params={
    "C":[1,5,10],
    "gamma":[.0001,.002,.01,.1]
}
grid=GridSearchCV(model_svc,params,verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_trains, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
